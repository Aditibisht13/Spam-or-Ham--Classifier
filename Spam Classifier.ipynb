```python
import pandas as pd
```

```python
df=pd.read_csv('spam.csv',encoding="ISO-8859-1")
df=df.rename(columns={'v1':'target','v2':'email'})
df=df[['email','target']]
```

```python
df
```

```python
df.info()
```

```python
import re
def clean(str_):
    str_=str_.lower()
    str_=re.sub('[^A-Za-z0-9]+',' ',str_)
    return str_
```

```python
clean("hey you HAVE a lottery of 100$")
```

```python
df['cleaned_email']=df['email'].apply(clean)
```

```python
df.head(3)
```

# Covert text to count frequencies

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
X_numeric = vectorizer.fit_transform(df['cleaned_email']).toarray() #converted text email to numerical feature
```

```python
Y=df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
vectorizer.vocabulary_
```

```python
Y = df['target']
```

```python
X_numeric[0]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_numeric,Y, test_size=0.2,shuffle=True)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB()
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
